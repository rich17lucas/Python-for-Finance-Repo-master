```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

```

```python
airline = pd.read_csv('airline_passengers.csv', index_col='Month')
airline.head()
```

```python
airline.plot()
```

```python
# Data clean up - remove missing values
airline.dropna(inplace=True)
airline.count()
```

```python
# Convert the index to a datetime index
airline.index = pd.DatetimeIndex(airline.index) 
```

```python
airline.head()
```

```python
# Start working with ETS (Error, Trend, Seasonality) analysis
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model='additive')
result.seasonal.plot()
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
result.seasonal.plot()
```
